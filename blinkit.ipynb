```python
plt.figure(figsize=(12, 5))
sns.countplot(data=df, y="category", order=df["category"].value_counts().index, palette="Blues_r")
plt.title("Product Count by Category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
top_brands = df["brand"].value_counts().head(10)
plt.figure(figsize=(12, 5))
sns.barplot(x=top_brands.values, y=top_brands.index, palette="coolwarm")
plt.title("Top 10 Brands by Product Count")
plt.xlabel("Number of Products")
plt.ylabel("Brand")
plt.show()
```

```python
# Price vs mrp
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="mrp", y="price", alpha=0.7, color="blue")
plt.plot([df["mrp"].min(), df["mrp"].max()], [df["mrp"].min(), df["mrp"].max()], "r--", label="MRP = Price")
plt.title("MRP vs Selling Price")
plt.xlabel("MRP (Max Retail Price)")
plt.ylabel("Selling Price")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["margin_percentage"], bins=30, kde=True, color="darkgreen")
plt.title("Distribution of Margin Percentage")
plt.xlabel("Margin (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["shelf_life_days"], bins=30, kde=True, color="purple")
plt.title("Shelf Life Distribution (Days)")
plt.xlabel("Shelf Life (Days)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=df[["min_stock_level", "max_stock_level"]], palette="Set2")
plt.title("Min vs Max Stock Level Distribution")
plt.ylabel("Stock Level")
plt.xticks(ticks=[0, 1], labels=["Min Stock Level", "Max Stock Level"])
plt.show()
```

```python
# Calculate discount and discount percentage
df['discount'] = df['mrp'] - df['price']
df['discount_percentage'] = (df['discount'] / df['mrp']) * 100

# Group by category to see average discount and margin
elasticity = df.groupby('category')[['discount_percentage', 'margin_percentage']].mean().reset_index()
print(elasticity)

# Scatter plot to explore relationship
plt.figure(figsize=(8, 5))
sns.scatterplot(x='discount_percentage', y='margin_percentage', size='price', hue='category', data=df)
plt.title('Discount Percentage vs Margin Percentage by Category')
plt.show()
```

```python
# Calculate average stock level
df['avg_stock_level'] = (df['min_stock_level'] + df['max_stock_level']) / 2

# Assume turnover is inversely related to shelf life and stock level (simplified)
df['turnover_proxy'] = 1 / (df['avg_stock_level'] * df['shelf_life_days'])

# Summary by category
turnover_summary = df.groupby('category')['turnover_proxy'].mean().reset_index()
print(turnover_summary)

# Visualize turnover by category
plt.figure(figsize=(10, 5))
sns.barplot(x='turnover_proxy', y='category', data=turnover_summary.sort_values('turnover_proxy', ascending=False))
plt.title('Estimated Inventory Turnover by Category')
plt.show()
```

```python
# Group by brand
brand_summary = df.groupby('brand').agg({
    'product_id': 'count',
    'price': 'mean',
    'margin_percentage': 'mean',
    'shelf_life_days': 'mean'
}).reset_index().rename(columns={'product_id': 'product_count'})
print(brand_summary.sort_values('margin_percentage', ascending=False).head(10))

# Visualize top brands by margin
top_brands = brand_summary.nlargest(10, 'margin_percentage')
plt.figure(figsize=(10, 5))
sns.barplot(x='margin_percentage', y='brand', data=top_brands)
plt.title('Top 10 Brands by Average Margin Percentage')
plt.show()
```

```python
# Bin shelf life into categories
df['shelf_life_category'] = pd.cut(df['shelf_life_days'], bins=[0, 30, 90, 365, float('inf')], labels=['Short (<30)', 'Medium (30-90)', 'Long (90-365)', 'Very Long (>365)'])

# Group by shelf life category
shelf_life_summary = df.groupby('shelf_life_category')[['price', 'margin_percentage']].mean().reset_index()
print(shelf_life_summary)

# Visualize price by shelf life category
plt.figure(figsize=(8, 5))
sns.barplot(x='shelf_life_category', y='price', data=shelf_life_summary)
plt.title('Average Price by Shelf Life Category')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

# Select features for anomaly detection
features = ['price', 'margin_percentage', 'shelf_life_days']
X = df[features]

# Fit Isolation Forest
iso_forest = IsolationForest(contamination=0.1, random_state=42)
df['anomaly'] = iso_forest.fit_predict(X)

# Filter anomalies
anomalies = df[df['anomaly'] == -1]
print(anomalies[['product_name', 'category', 'brand', 'price', 'margin_percentage', 'shelf_life_days']])

# Visualize anomalies
plt.figure(figsize=(8, 5))
sns.scatterplot(x='price', y='margin_percentage', hue='anomaly', data=df, palette={1: 'blue', -1: 'red'})
plt.title('Anomalies in Price vs Margin Percentage')
plt.show()
```
